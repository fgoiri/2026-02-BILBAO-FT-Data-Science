## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
'''
Mi código es bastante básico. 
No he sabido guardar los resultados y que si los pides te los devuelva. He creado la lista para que los guarde,
pero luego no se recuperarlos. Y luego tampoco está a prueba de meter otros valores que no sean las operaciones.
'''

# Primero hacemos la variable resultado en una lista para guardar los resultados.
resultado = []
print("BIENVENIDO A LA CALCULADORA")
# Primero en el bucle preguntamos la operacion que queremos usar
while True:
    print("¿Qué operación quiere realizar?\n Opciones disponibles:\n"
          " Sumar\n Restar\n Multiplicar\n Dividir\n Potencia\n Resultados\n Salir")
    operacion = input("Elige una opción: ").lower()
    
    if operacion == "salir":
        print("Gracias por usar nuestra calculadora")
        break
    
    elif operacion != "salir":
        print("Introduzca el primer número, por favor")
        num_1 = float(input())
        
        print("Introduzca el otro número, por favor")
        num_2 = float(input())
    
        if operacion == "sumar":
            resultado = num_1 + num_2
            print("El resultado de la suma es:", resultado)
            
        elif operacion == "restar":
            resultado = num_1 - num_2
            print("El resultado de la resta es:", resultado)
            
        elif operacion == "multiplicar":
            resultado = num_1 * num_2
            print("El resultado de la multiplicación es:", resultado)
            
        elif operacion == "dividir":
            if num_2 == 0:
                print("No se puede dividir entre 0")
            else:
                resultado = num_1 / num_2
                print("El resultado de la división es:", resultado)
            
        elif operacion == "potencia":
            resultado = num_1**num_2
            print("El resultado de la potencia es:", resultado)
        elif operacion == "resultados":
            print("Los resultados de las operaciones realizadas son:", resultado)
```
