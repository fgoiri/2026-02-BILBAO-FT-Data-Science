## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

$$radianes = grados*(pi/180)$$

````python
import math
````

````python
def conversor(grados):
    radianes = grados * (math.pi/180)
    return radianes

conversor(180)
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
print("Introduzca un número por favor")
num_1 = float(input())
print("Introduzca otro número por favor")
num_2 = float(input())
suma = num_1 + num_2
if suma == 10:
    print("La suma es igual a 10, el total es:", suma)
elif suma > 10:
    print("La suma es mayor a 10, el tota es:", suma)
else:
    print("La suma es inferior a 10, el total es:", suma)
````

## Ejercicio 8
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

Compruébalo después con python

1. Como uno de los valores es False, el resultado de la operación es `False`
2. Hay que hacer primero los paréntesis. El resultado de ambos paréntesis es True. La siguiente operación es True and True, y su resultado es True. Por último hacemos el not True, por lo tanto el resultado es `False`.
3. Primero hay que hacer la operación entre paréntesis, y el resultado de esta es True. Después hay que hacer la operación True and False, y su resultado es `False`.
4. Primero hacemos los not, por lo tanto la operación quedará: True and True and False. El resultado de esta operación es `False`.

````python
print(True and True and False)
print(not ((True or False) and (True or False)))
print((False or False or False or True) and False)
print(not False and True and not True)
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "Lo primero de todo, cómo están los máquinas"

1. Pasarlo todo a mayúsculas
2. Pasarlo todo a minúsculas
3. Pasar a mayúsculas solo las iniciales de las palabras
4. Crea una lista dividiendolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
bisbi = "Lo primero de todo, cómo están las máquinas"
print(bisbi.upper())
print(bisbi.lower())
print(bisbi.title())
lista = bisbi.split(" ")
print(lista)
texto_sin_comas = bisbi.replace(",",";")
print(texto_sin_comas)
texto_sin_a = bisbi.replace("a","").replace("á","")
print(texto_sin_a)
````
