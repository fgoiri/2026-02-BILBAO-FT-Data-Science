## Ejercicio 4
Realiza un conversor de grados Celsius a Fahrenheit. El conversor tiene que recibir el input del usuario y devolver la conversión.

La conversión de grados sigue la siguiente fórmula: F = C × (9 / 5) + 32 

```python
print("Escribe los grados Celsius que quieras convertir a Fahrenheit. Recuerde que los decimales debe introducirlo con un punto")
celsius = float(input())
fahr = celsius * (9/5) + 32
print(round(fahr, 2))
```
